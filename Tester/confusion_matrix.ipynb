```python
import numpy as np
import tensorflow as tf
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os
class_mapping_4_class = {
    0: 'Bike',
    1: 'BUS',
    2: 'Common',
    3: 'Truck'
}

class_mapping_4_class_2 = {
    0: 'BUS',
    1: 'Bike',
    2: 'Common',
    3: 'Truck'
}

class_mapping_4_class_3 = {
    0: 'Common',
    1: 'Bike',
    2: 'Truck',
    3: 'BUS'
}


model_paths = [
    ("../Jupyter/ResNet50/ResNet50.h5", (128, 128), class_mapping_4_class_2),
    ("../Jupyter/DenseNet121/DenseNet121.h5", (128, 128), class_mapping_4_class_3),

    ("../Jupyter/MobileNet/MobileNet.h5", (128, 128), class_mapping_4_class),
    ("../Jupyter/MobileNet_change_freeze/MobileNet_change_freeze.h5", (128, 128), class_mapping_4_class),
    ("../Jupyter/MobileNet_change_freeze_class_weights/MobileNet_change_freeze_class_weights.keras", (128, 128), class_mapping_4_class_2),
    ("../Jupyter/MobileNet_change_freeze_2/MobileNet_change_freeze_2.keras", (128, 128), class_mapping_4_class_2),
    ("../Jupyter/MobileNet_change_freeze_3/MobileNet_change_freeze_3.keras", (128, 128), class_mapping_4_class_2),
    ("../Jupyter/MobileNet_change_freeze_3_early_stopping/MobileNet_change_freeze_3_early_stopping.keras", (128, 128), class_mapping_4_class_2),
    ("../Jupyter/MobileNet_no_weights/MobileNet_no_weights.h5", (128, 128), class_mapping_4_class),
    
    ("../Jupyter/EfficientNetB0/EfficientNetB0.h5", (224, 224), class_mapping_4_class),
    ("../Jupyter/cnn_4_class/cnn_4_class_2/cnn_4_class_2_5.h5", (128, 128), class_mapping_4_class),
    ("../Jupyter/cnn_4_class/cnn_4_class_2/cnn_4_class_2_6.h5", (128, 128), class_mapping_4_class),
]

validation_dir = '../Dataset_4_class/Test'
validation_datagen = ImageDataGenerator(rescale=1.0 / 255)

validation_generator = validation_datagen.flow_from_directory(
    validation_dir,
    target_size=(128, 128),  #   
    batch_size=32,
    class_mode='categorical',
    shuffle=False  #      
)

#       
def plot_confusion_matrix_and_metrics(y_true, y_pred, class_names, model_name):
    # class_names   
    if not isinstance(class_names, list) or not all(isinstance(c, str) for c in class_names):
        raise ValueError("class_names must be a list of strings.")

    #   
    conf_matrix = confusion_matrix(y_true, y_pred)

    # Classification Report 
    report = classification_report(y_true, y_pred, target_names=class_names, output_dict=True)

    # Classification Report  
    report_text = classification_report(y_true, y_pred, target_names=class_names)

    #  
    plt.figure(figsize=(10, 8))

    #   
    plt.subplot(2, 1, 1)
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title(f'Confusion Matrix: {model_name}')

    #    
    plt.subplot(2, 1, 2)
    plt.text(0.01, 0.05, str(report_text), {'fontsize': 10}, fontfamily='monospace')  # fontfamily='monospace' for alignment
    plt.axis('off')  #  


    #    
    folder_path = './confusion_matrix'

    #   
    os.makedirs(folder_path, exist_ok=True)

    #   
    plt.tight_layout()
    plt.savefig(f'./{folder_path}/{model_name}_confusion_matrix_and_metrics.png')
    plt.show()

#         
for model_path, target_size, class_mapping in model_paths:
    print(f"Loading model: {model_path}")
    
    #  
    model = tf.keras.models.load_model(model_path)
    
    #       
    validation_generator = validation_datagen.flow_from_directory(
        validation_dir,
        target_size=target_size,  #   
        batch_size=32,
        class_mode='categorical',
        shuffle=False
    )
    
    # Provided Class Mapping 
    print("Provided Class Mapping:", class_mapping)

    # `y_true` `Provided Class Mapping`  
    y_true = []
    for idx in validation_generator.filenames:
        # Extract class name from filename path
        class_name = idx.split('/')[0]  #     
        y_true.append(list(class_mapping.keys())[list(class_mapping.values()).index(class_name)])
    
    #    
    class_names = list(class_mapping.values())  # ['Common', 'Bike', 'Truck', 'Bus']

    #  
    y_pred = model.predict(validation_generator)
    y_pred_classes = np.argmax(y_pred, axis=1)

    #        
    plot_confusion_matrix_and_metrics(y_true, y_pred_classes, class_names, model_path.split('/')[-1])
```
