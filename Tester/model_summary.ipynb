```python
# MobileNet 모델 구성
import tensorflow as tf


base_model = tf.keras.applications.MobileNet(
    input_shape = (128, 128, 3),  # 입력 이미지 크기
    include_top = False,  # 최상위 분류 레이어 제외
    weights = 'imagenet'  # ImageNet 가중치 사용
)

from tensorflow.keras.models import load_model

model_path = '../Jupyter/MobileNet_change_freeze_3_early_stopping/MobileNet_change_freeze_3_early_stopping.keras'
model = load_model(model_path)

# Base Model Summary
print("Base Model Summary:")
base_model.summary()

# 전체 모델 Summary
print("\nEntire Model Summary:")
model.summary()
```
